```python
df_anime['type'].fillna(df_anime['type'].mode(), inplace=True)
```

```python
dum_var=df_anime['genre'].str.get_dummies(sep=',')
dum_var.shape
```

```python
dum_var.fillna(dum_var.median(),inplace=True)
```

```python
dum_var.head()
```

```python
print(df_anime.dtypes)
print(df_rating.dtypes)
```

```python
df_rating['rating'].value_counts()
```

```python
anime_wt_genre=df_anime.copy()
```

```python
print(anime_wt_genre.isnull().any())
print(dum_var.isnull().any())
```

```python
anime_wt_genre=pd.concat([anime_wt_genre,dum_var],axis=1)
anime_wt_genre.head()
```

```python
userinput=[{'name':'Kimi no Na wa.','rating':10},
          {'name':'Fullmetal Alchemist: Brotherhood','rating':8},
          {'name':'Gintama°','rating':8},
          {'name':'Steins;Gate','rating':7}]
```

```python
inanime=pd.DataFrame(userinput)
inanime.head()
```

```python
inid=df_anime[df_anime['name'].isin(inanime['name']).tolist()]
inanime=pd.merge(inid,inanime)
inanime.head()
```

```python
inanime=inanime.drop('genre',1).drop('type',1).drop('episodes',1).drop('members',1)
```

```python
inanime.head()
```

```python
useranim=anime_wt_genre[anime_wt_genre['anime_id'].isin(inanime['anime_id'].tolist())]
useranim.head()
```

```python
usergenre=useranim.drop('anime_id',1).drop('name',1).drop('genre',1).drop('type',1).drop('episodes',1).drop('members',1)
usergenre.head()
```

```python
userprof=usergenre.transpose().dot(inanime['rating'])
userprof.head()
```

```python
recomm=((dum_var*userprof).sum(axis=1)/(userprof.sum()))
```

```python
recomm=recomm.sort_values(ascending=False)
```
